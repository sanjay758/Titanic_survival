```python
def Embarked(Embarked):
    if   Embarked=='S':
        return  "Southampton"
    elif Embarked=='Q':
        return "Quennstown"
    else:
        return "Cherbourg"
data['New_Embarked']=data['Embarked'].apply(Embarked)
data
```

```python
dup=data.duplicated().any()
print("Any duplicate Value? ",dup)
```

```python
data.drop('Cabin',axis=1,inplace=True)
data
```

```python

df_Titanic=data[['Survived', 'Pclass', 'Name', 'Gender_new', 'Age','SibSp', 'Parch', 'Fare', 'Embarked']]
df_Titanic
```

```python
df_Titanic=pd.get_dummies(df_Titanic,columns=['Embarked'])
df_Titanic
```

```python
def Result(Result):
    if Result==0:
        return "Died"
    else :
        return "Survived"
data['PassengerCase']=data['Survived'].apply(Result)
data
```

```python
def classes(clas):
    if clas==1:
        return "First Class"
    elif clas==2:
        return "Second Class"
    else:
        return "Third Class"
data['New_Pclass']=data['Pclass'].apply(classes)
data
    
```

```python
def age_cat(age):
    if age>=80.0:
        return "Elderly"
    elif age>=66.0:
        return "Middle_Age"
    elif age>=26.0:
        return "youth"
    
    elif age>=18.0:
        return "Adolescence"
    elif age>=11.0:
        return "underage"
    else:
        return "Kides"
data['Age_Cat']=data['Age'].apply(age_cat)
data
    
```
